```python
import os
from keras.preprocessing import image
import random
from keras import utils
import numpy as np

from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
```

```python
folders = os.listdir('datasets/pokemon')
print(folders)
```

```python
image_data = []
labels = []
count = 0

for folder in folders:
    path = os.path.join("datasets/pokemon",folder)
    for file in os.listdir(path):
        try:
            img = image.load_img(os.path.join(path,file),target_size=(224,224))
            img_array = image.img_to_array(img)
            image_data.append(img_array)
            labels.append(count)
        except Exception as e:
            print(e)
    count += 1
```

```python

combined_dataset = list(zip(image_data,labels))
random.shuffle(combined_dataset)
image_data[:], labels[:] = zip(*combined_dataset)
```

```python


X_train = np.array(image_data)
Y_train = np.array(labels)

Y_train = utils.to_categorical(Y_train)
```

```python
model = ResNet50(include_top = False, weights = 'imagenet', input_shape=(224,224,3))

print(model.summary())
```

```python
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256, activation = 'relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(10,activation = 'softmax')(d1)
```

```python
model_new = Model(inputs = model.input, outputs = fc2)
model_new.summary()
```

```python
from keras.applications.resnet50 import preprocess_input


image_path = 'datasets/pokemon/Pikachu/e8e8119e227246568e4b03db68ea5e8d.jpg'

img = image.load_img(image_path, target_size=(224,224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)

pred = model_new.predict(x)
print(np.argmax(pred))
#wrong output so we need to fine-tune
```

```python
adam = Adam(learning_rate=0.00003)
model_new.compile(loss = 'categorical_crossentropy', optimizer = adam, metrics = ['accuracy'])
```

```python
for each_layer in range(len(model_new.layers)):
    print(each_layer, model_new.layers[each_layer])
```

```python
for i in range(169):
    model_new.layers[i].trainable = False

print(model_new.summary())
```

```python
hist = model_new.fit(X_train, Y_train, shuffle = True, batch_size = 16, epochs = 8, validation_split = 0.20)

```
