```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('datasets/iris.csv')
X = dataset.iloc[0:100,[0,1,2,3]].values
Y = dataset.iloc[0:100,4].values

Y = np.where(Y == 'Iris-setosa', 0, 1)

weights = np.array([0.0,0.0,0.0,0.0])
bias = 0.0
```

```python
print("X:\n",X)
print("Y:\n",Y)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def forward_propogation(X,W,b):
    weighted_sum = np.dot(X,W)+b
    prediction = sigmoid(weighted_sum)
    return prediction

def gradient(X,Y,Y_predicted):
    '''Gradient of Weights and Bias'''
    Error = Y_predicted - Y
    dW = np.dot(X.T,Error)
    db = np.sum(Error)
    return dW,db

def update_parametes(W,b,dW,db,learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return W,b

def train(X,Y,W,b):
    epochs = 1000
    learning_rate = 0.1

    for i in range(epochs):
        Y_predicted = forward_propogation(X,W,b)
        dW,db = gradient(X,Y,Y_predicted)
        W,b = update_parametes(W,b,dW,db,learning_rate)
        

    return W,b

```

```python
#Model Training
W,b = train(X,Y,weights,bias)
```

```python
print("Weights: ",W)
print("Bias: ",b)
```

```python
#predicting value
print("Target Value")
print(Y)
A2 = forward_propogation(X,W,b)

print("Predicted Value")
Y_predicted = (A2 > 0.5) * 1
print(Y_predicted)
```

```python
#Compare predicted and actual labels

comparision = Y_predicted == Y
equal_arrays = comparision.all()
print("Prediction Accuracy: ",equal_arrays)
```
